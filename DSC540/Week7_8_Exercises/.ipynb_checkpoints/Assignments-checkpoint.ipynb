```python
# Import the pandas, numpy and matplotlib libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# This is to ignore any warnings on output.
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset
candyData = pd.read_excel('BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx')
candyData.head(10)
```

```python
## checking datatypes
candyData.dtypes
```

## Chapter 7

```python
## Remove duplicates
candyData2016 = candyData.drop_duplicates()
candyData2016.shape
```

```python
# Filter out missing observations
# Keep only the rows with at least 2 non-NA values.
df = candyData2016.dropna(axis=0, how='all', thresh=70)
df.shape
```

```python
print("Old dataframe Rec Cnt:", len(candyData2016), "\nNew dataframe Rec Cnt:", 
       len(df), "\nNumber of rows with at least 1 NA value: ",
       (len(candyData2016)-len(df)))
```

```python
# Fill in missing data
FillMissingData_df = df.fillna(method='bfill')
FillMissingData_df.head()
```

```python
# 4) Replace values to make data consistant
FillMissingData_df['Which country do you live in?'].head()
```

```python
FillMissingData_df['Which country do you live in?'] = FillMissingData_df['Which country do you live in?'].apply(lambda x: 'USA' if (x =='usa') | (x == 'US') 
                                                             else x)
candyData2016_df = FillMissingData_df.copy() ;
candyData2016_df['Which country do you live in?'].head()

```

## Chapter 8

```python
## create hierarchical index
## candyData2016_df.head()
try:
    hc_candyData2016_df = candyData2016_df.copy() ; 
    hc_candyData2016_df.set_index(['Timestamp','Your gender:','How old are you?','Which country do you live in?'], inplace=True)
    hc_candyData2016_df.sort_index(inplace=True)
    print(hc_candyData2016_df.head(5))
    print(hc_candyData2016_df.tail(5))
except KeyError:
    print(hc_candyData2016_df.head(5))
    print(hc_candyData2016_df.tail(5))
```

```python
## Merge datasets

candyData2015 = pd.read_excel('CANDY-HIERARCHY-2015-SURVEY-Responses.xlsx', sheet_name='Form Responses 1')

df_merged = pd.merge(candyData2016_df, candyData2015, how='left')
df_merged.head(3)
```

```python
#2 Pivot the data
pivot_candyData2015_df = candyData2015.copy() ;
p_t = pd.pivot_table(pivot_candyData2015_df , values='Guess the number of mints in my hand.', index=['Timestamp' , 'Are you going actually going trick or treating yourself?'],
                    columns=['How old are you?'], aggfunc=np.sum)
p_t
```

## Chapter 10

```python
## Grouping with functions

candyData2016_df2 = candyData2016_df[candyData2016_df['How old are you?'].str.match('^\d+$',na=True)]
candyData2016_df3 = candyData2016_df2[candyData2016_df2['Which country do you live in?'] == 'USA']
candyData2016_df3
```

```python
agg_candyData2016df = candyData2016_df3.copy()
grouped = agg_candyData2016df.groupby(['Your gender:'])
grouped_pt = grouped['How old are you?']
grouped_pt.agg(['min', 'max'])
```

```python
## Grouping with index level

candyData2016_df3.dropna(axis=0, how='all')
candyData2016_df3['Betty or Veronica?'] = candyData2016_df3['Betty or Veronica?'].apply(lambda x: 'USA' if (x =='???') | (x == 'Dont Know yet') 
                                                             else x)
candyData2016_df4 = candyData2016_df3[['Betty or Veronica?',
                            'Your gender:','How old are you?']]
candyData2016_df4.set_index(['Betty or Veronica?',
                          'Your gender:','How old are you?'], inplace=True)
candyData2016_df4.sort_index(inplace=True)
candyData2016_df4
```
