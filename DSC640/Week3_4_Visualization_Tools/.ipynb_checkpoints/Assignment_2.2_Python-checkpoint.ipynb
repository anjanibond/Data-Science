```python
# Import libraries
import csv
#import xlrd
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
# Read world population data
fileData1 = 'ex2-2/world-population.xlsm'
population = pd.read_excel(fileData1)
print(population.head())
```

#### World Population data
#### Line and Step Charts

```python
X = population['Year']
Y = population['Population']
plt.title('Python - World Population Growth - Line Chart')
plt.plot(X, Y)
plt.xlabel("Year")
plt.ylabel("World Population in Billions")
plt.show()
```

```python
plt.title('Python - World Population Growth - Step Chart')
plt.step(X, Y)
plt.xlabel("Year")
plt.ylabel("World Population in Billions")
plt.show()
```
